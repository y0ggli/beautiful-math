```python
print(text)
```

## Markov 1 Anker

```python
chain = {}
index = 1
for word in text[index:]:
    key = text[index - 1]
    if key in chain:
        chain[key].append(word)
    else:
        chain[key] = [word]
    index += 1

print(chain)    
```

```python
length = 100
word1 = random.choice(list(chain.keys()))
message = word1.capitalize()            

while len(message.split(' ')) < length:
    word2 = random.choice(list(chain[word1]))
    word1 = word2
    message += ' ' + word2
    
print(message)    
```

## Markov 2 Anker

```python
chain = {}
index = 2
for word in text[index:]:
    key = text[index - 2] + ' ' + text[index - 1]
    if key in chain:
        chain[key].append(word)
    else:
        chain[key] = [word]
    index += 1

print(chain)    
```

```python
length = 100
word1 = random.choice(list(chain.keys()))
message = word1.capitalize()            
#print(word1)
while len(message.split(' ')) < length: 
        newKey = word1.split(' ')
        #print(newKey)
        try:
            word2 = newKey[1] + ' ' + random.choice(list(chain[word1]))
        except KeyError:
            print("no key")
        #print(word2)
        word1 = word2
        message += ' ' + word2.split(' ')[1]

message += "."        
print(message)
```
