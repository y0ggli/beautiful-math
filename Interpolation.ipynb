```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
import math

%matplotlib inline
%matplotlib notebook
```

```python
seedArray = np.random.uniform(low=0.0, high=1.0, size=(10))
print(seedArray)
```

```python
def linear(x):
    global seedArray
    x1 = math.floor(x)
    x2 = math.ceil(x)
    y1 = seedArray[x1]
    y2 = seedArray[x2]
    y = (y1*(1-(x-x1)) + y2*(x-x1))
    #y = y1 + ((y2-y1)/(x2-x1))*(x-x1)
    return y

print(linear(2.5))
```

```python
def smoothstep(x):
    global seedArray
    x1 = math.floor(x)
    x2 = math.ceil(x)
    y1 = seedArray[x1]
    y2 = seedArray[x2]
    
    x = x-x1
    tempy = 3*(x**2)-2*(x**3)
    y = tempy*(y2-y1)+y1
    return y

print(smoothstep(2.5))
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,9,1000)

ax.plot(x, np.vectorize(linear)(x))
ax.plot(x, np.vectorize(smoothstep)(x))

plt.show()
```

```python
width = 200;
height = 100;
seedArray2D = np.random.uniform(low=0.0, high=1.0, size=(height, width))
#print(seedArray2D)
```

```python
'''
def getColor(grid):
    colorArray = np.empty((width, height), dtype=(object))
    for x in range(width):
         for y in range(height):
                    r = grid[x][y]
                    g = grid[x][y]
                    b = grid[x][y]
                    colorArray[x][y] = (r,g,b)
    return colorArray
'''

def smoothstep(p1,p2,p):
    temp = 3*(p**2)-2*(p**3)
    pnew = temp*(p2-p1)+p1
    return pnew

def valueNoise2D(octave):
    global seedArray2D
    global width
    global height
    
    grid = np.zeros((height, width), dtype='float')
    period = 1<<octave
    frequency = 1.0 / period
    for x in range(width):
        x0 = (x // period) * period
        x1 = (x0 + period) % width
        horizontal_blend = (x-x0) * frequency
        #print("X:"+str(x))
        #print("X0:"+str(x0))
        #print("X1:"+str(x1))
        #print("blend:"+str(horizontal_blend))
        for y in range(height):
            y0 = (y // period) * period
            y1 = (y0 + period) % height
            vertical_blend = (y-y0) * frequency
            top = smoothstep(seedArray2D[y0][x0],seedArray2D[y0][x1], horizontal_blend)
            bottom = smoothstep(seedArray2D[y1][x0],seedArray2D[y1][x1], horizontal_blend)
            grid[y][x] = smoothstep(top, bottom, vertical_blend)
            
    return grid

def generateNoise(octaveCount):
    global width
    global height
    
    smoothNoise = np.zeros((octaveCount, height, width), dtype='float')
    persistance = 0.7
    
    for i in range(octaveCount):
        smoothNoise[i] = valueNoise2D(i)
    
    
    newValueNoise = np.zeros((height, width), dtype='float')
    amplitude = 1.0
    totalAmplitude = 0.0
    
    for o in range(octaveCount-1,0,-1):
        amplitude *= persistance
        totalAmplitude += amplitude
        for x in range(width):
            for y in range(height):
                newValueNoise[y][x] += smoothNoise[o][y][x] * amplitude 
    
    for x in range(width):
        for y in range(height): 
            if totalAmplitude > 0:
                newValueNoise[y][x] /= totalAmplitude    
    return newValueNoise
                
#print(valueNoise2D(4))    
#print(generateNoise(4))
#img = valueNoise2D(4)
img = generateNoise(6)
plt.imshow(img, cmap='gray',clim=(0,1))
plt.show()



```
